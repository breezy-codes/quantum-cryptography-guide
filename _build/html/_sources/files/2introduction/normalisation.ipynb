# **Normalisation**

For probabilities to be valid and sum up to 1, every quantum state $|\psi\rangle$ must become normalized. **Normalisation** ensures the total probability of measuring all possible outcomes is 1.

If we have:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$

Then we also have:

$$|\alpha|^2 + |\beta|^2 = 1$$

The figure below represents the range of possible values $\alpha$ and $\beta$ can take, such as their squared magnitudes sum to 1. This emphasizes that as the magnitude of one coefficient increases, the magnitude of the other must decrease to maintain Normalisation.

```{admonition} Explanation
:class: tip, dropdown

**What is Normalisation?**

Normalisation in quantum mechanics ensures that the total probability of all possible outcomes of a measurement is 1 (or 100%). This is crucial because it reflects the fact that the quantum system must be in one of the possible states when measured.

**Mathematical Representation:**

If we have a quantum state expressed as:

$$ |\psi\rangle = \alpha|0\rangle + \beta|1\rangle $$

Here, $|\psi\rangle$ is a combination of the states $|0\rangle$ and $|1\rangle$ with probability amplitudes $\alpha$ and $\beta$, respectively.

**The Normalisation Condition:**

The Normalisation condition is given by:

$$ |\alpha|^2 + |\beta|^2 = 1 $$

This equation means that the sum of the squared magnitudes of the probability amplitudes must equal 1. It ensures that the probabilities of measuring the quantum state in $|0\rangle$ or $|1\rangle$ add up to 1.

**In simpler terms:**

Imagine you have a quantum system that can be in either state $|0\rangle$ or state $|1\rangle$. The coefficients $\alpha$ and $\beta$ tell you how likely the system is to be in each state. The Normalisation condition ensures that these likelihoods are balanced so that the total probability is always 100%.

In other words, if $\alpha$ gets larger, $\beta$ must get smaller, and vice versa, to keep the sum of their squared magnitudes equal to 1. This balancing act is crucial for maintaining the integrity of probability calculations in quantum mechanics.
```

````python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

# Plot the circle
plt.figure(figsize=(8, 8))
plt.plot(x, y, label=r"$|\alpha|^2 + |\beta|^2 = 1$")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("Normalization Constraint in Quantum Mechanics")
plt.xlabel(r"$|\alpha|$")
plt.ylabel(r"$|\beta|$")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
````

## **Example of Normalisation**

Let's take an example to illustrate Normalisation. Suppose we have a quantum state:

$$|\psi\rangle = \frac{1}{\sqrt{3}}|0\rangle + \sqrt{\frac{2}{3}}|1\rangle$$

We need to check if this state is normalised.

**Step 1:** Calculate the squared magnitudes of the coefficients.

$$|\alpha|^2 = \left|\frac{1}{\sqrt{3}}\right|^2 = \frac{1}{3}$$

$$ |\beta|^2 = \left|\sqrt{\frac{2}{3}}\right|^2 = \frac{2}{3}$$

**Step 2:** Sum the squared magnitudes.

$$ |\alpha|^2 + |\beta|^2 = \frac{1}{3} + \frac{2}{3} = 1 $$

Since the sum of the squared magnitudes is 1, the state $|\psi\rangle$ is normalised.

We check if it's normalised by adding up the squares of the coefficients. If they add up to 1, the state is normalised. In this case, they do, so our state is correctly normalised.
