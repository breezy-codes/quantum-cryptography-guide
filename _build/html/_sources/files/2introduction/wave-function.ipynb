# **The Wave Function Collapse**


## **Observer Effect and Collapse**

Thinking back to our coin analogy previously on superposition, this is the concept where a qubit can exist in a combination of states. However, upon measurement, it collapses into one of those possible states; this observation is known as the observer effect.

For a qubit:
$$\ket{q} = \alpha\ket{0} + \beta\ket{1} $$

If $\ket{q}$ is measured to be in $\ket{0}$, the subsequent measurements will consistently return $\ket{0}$. This is known as the **wave function collapse**.

## **Wave Function Collapse**

The wave function provides a description of the quantum state of a system. Before measurement, the function generally describes a superposition of several possible states. Once measured, the wave function “collapses” to a singular state, which is the result of that measurement.

This phenomenon is rather mysterious as it contrasts with classical physics. In classical physics, measurements will simply reveal properties that the system already had. However, in quantum physics, measurements can change the state of the system.

For a qubit described by the state $\ket{q} = \alpha\ket{0} + \beta\ket{1}$, where $|\alpha|^2$ gives the probability of finding the qubit in state $\ket{0}$ and $|\beta|^2$ gives the probability of finding the qubit in state $\ket{1}$, by doing the following:
1. Collapses the state to $\ket{0}$ with a probability $|\alpha|^2$.
2. Collapses the state to $\ket{1}$ with a probability $|\beta|^2$.

Subsequent measurements will now yield the same results. In the following example, the graph is set to measuring the probability of $\ket{1}$.

```{admonition} Explanation
:class: tip, dropdown

**Observer Effect:**

The observer effect in quantum mechanics refers to the fact that the act of measuring a quantum system affects the system. Before measurement, a quantum system can exist in a superposition of states. However, once measured, it collapses into one of the possible states.

**Wave Function:**

The wave function is a mathematical description of the quantum state of a system. It contains all the information about the system's state and can describe multiple possible outcomes simultaneously through superposition.

**Wave Function Collapse:**

When a measurement is made, the wave function collapses to a single state. This means that the superposition of states reduces to one state, and subsequent measurements will yield the same result. This collapse is what makes quantum mechanics different from classical physics, where measurements do not alter the state of the system.
```

````python
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.7
beta = (1 - abs(alpha)**2)**0.5

if beta.imag == 0:
    beta = beta.real

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x)

outcome = np.random.choice([0, 1], p=[abs(alpha)**2, abs(beta)**2])

plt.figure(figsize=(10, 6))

if outcome == 0:
    plt.fill_between(x, y, where=((x > 1) & (x < 2)), color='lightblue', label='Probability of $|0⟩$')
    plt.vlines(x=1.5, ymin=0, ymax=1, colors='b', linestyles='dashed', label='Collapse to $|0⟩$')
else:
    plt.fill_between(x, y, where=((x > 4) & (x < 5)), color='lightgreen', label='Probability of $|1⟩$')
    plt.vlines(x=4.5, ymin=0, ymax=-1, colors='g', linestyles='dashed', label='Collapse to $|1⟩$')

plt.plot(x, y, label='Superposition State')
plt.title('Visualization of Wave Function Collapse')
plt.xlabel('Possible States')
plt.ylabel('Probability Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````


```{prf:algorithm} Example: Measuring Probability of State $\ket{1}$
:label: alg-collapse

**Given Quantum State:** $|q\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle$

**Objective:** Determine the probability of measuring the state $|1\rangle$ in the given quantum state $|q\rangle$.

**Step 1:** Calculate the inner product $\langle 1|q\rangle$ to find the amplitude of $|1\rangle$ in $|q\rangle$.

*Inner product calculation:*

$$\langle 1|q\rangle = \frac{1}{\sqrt{2}}\langle 1|0\rangle + \frac{1}{\sqrt{2}}\langle 1|1\rangle $$

*Expanding using the given state.*

**Recall:** Orthogonality of basis states implies $\langle 1|0\rangle = 0$ and $\langle 1|1\rangle = 1$. Thus,

$$\langle 1|q\rangle = \frac{1}{\sqrt{2}}.$$

**Step 2:** Calculate the probability of measuring $|1\rangle$ by squaring the magnitude of the amplitude.
*Probability calculation:*

$$ p(1) = |\langle 1|q\rangle|^2 = \left|\frac{1}{\sqrt{2}}\right|^2 = \frac{1}{2}.$$

**Conclusion:** The probability of measuring the state $|1\rangle$ in the quantum state $|q\rangle$ is $0.5$ or 50%.
```

### Explanation

This example illustrates how to calculate the probability of measuring a specific state for a given quantum state. We start with a quantum state in superposition and use the inner product to find the amplitude of the basis state we are interested in. We then square the magnitude of this amplitude to get the probability of measuring the state. In this case, the state $|q\rangle$ is a superposition of $|0\rangle$ and $|1\rangle$, and we calculate the probability of measuring $|1\rangle$.
