# **How do computers classify algorithms?**

In classical computers, an algorithm is classified on how the resources used grow with the size of the input, this is known as an **algorithm’s complexity**. Complexity is expressed as a useful metric function, often using Big O notation, which provides an upper bound performance, ignoring constant factors and lower order terms.

**Time complexity** measures the amount of computational time an algorithm takes to compute as a function of the size n of the input. An algorithm with a time complexity of $O(n)$ is linear, meaning the time it takes to finish grows linearly with the size n of the input. An algorithm with a time complexity of $O(n^2)$ is quadratic, meaning the time it takes to finish grows quadratically with the size n of the input.

Before exploring how quantum computing is working towards changing the classification of algorithms, we must first understand why. Quantum computing is built on quantum physics, which of course is built on physics. Classical physics provides a framework for understanding the universe on a macroscopic scale, encompassing most things we observe in day-to-day life. However, when scientists looked at physics from a microscopic atomic and subatomic scale, they discovered physics couldn’t accurately predict or explain what was observed.

This led to the development of **quantum mechanics**, a branch of physics whose purpose is to describe the fundamental behaviour of matter and energy on incredibly small scales. Quantum mechanics wasn’t a replacement for physics, but more so a different approach for microscopic level. As the understanding of quantum mechanics grew, computer scientists and engineers realised the potential which led to the invention of quantum computing.

Quantum computers are rapidly improving, although they aren’t quite there yet when it comes to changing how algorithms are classified. This is due to the limited number of qubits and their noise (think interference) to reach **quantum advantage**, which refers to as the point where a quantum computer can solve a problem more efficiently than a classical computer.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## **Complexity Classes**

This graph illustrates different algorithmic complexity classes by plotting the relationship between input size and the number of operations or time required. The complexities include constant, linear, quadratic, and logarithmic, showcasing how they scale with increasing input size.

```python
n = np.linspace(1, 20, 400)
n = n[n > 0]
constant = [1 for _ in n]
linear = n
quadratic = n**2
logarithmic = np.log(n)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n, constant, label="Constant Time", lw=2)
plt.plot(n, linear, label="Linear", lw=2)
plt.plot(n, quadratic, label="Quadratic", lw=2)
plt.plot(n, logarithmic, label="Logarithmic", lw=2)
plt.legend(loc="upper left")
plt.xlabel("Input Size n")
plt.ylabel("Operations (or) Time")
plt.title("Algorithmic Complexity Classes", fontweight="bold")
plt.grid(True)
plt.yscale("log")
plt.show()
```

## **Factorisation Complexity**

This graph compares the estimated time required for factorising numbers of varying bit lengths using classical and quantum computers. The classical time is modeled exponentially, while the quantum time prediction follows a cubic relationship.

```python
n = np.linspace(2, 500, 300)
classical_time = np.exp(0.055 * n)
quantum_time = n**3

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n, classical_time, label="Classical Computer", lw=2)
plt.plot(n, quantum_time, label="Quantum Computer (Future Prediction)", lw=2, linestyle="--")

plt.legend()
plt.xlabel("Bit Length of Number to Factorize")
plt.ylabel("Time (Arbitrary Units)")
plt.title("Comparison of Factorization Time: Classical vs. Quantum", fontweight="bold")
plt.yscale("log")
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```
